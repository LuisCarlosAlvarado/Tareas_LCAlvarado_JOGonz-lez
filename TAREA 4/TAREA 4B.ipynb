### Ejemplo binomial: 
La distribución binomial modela el número de éxitos de n ensayos independientes donde hay una probabilidad p de éxito en cada ensayo.

Generar una variable aletoria binomial con parámetros $n=10$ y $p=0.7$. Recordar que
$$X\sim binomial(n,p) \longrightarrow p_i=P(X=i)=\frac{n!}{i!(n-i)!}p^i(1-p)^{n-i},\quad i=0,1,\cdots,n$$
> ## <font color ='red'>Tarea: Demostrar la validez de la siguiente ecuación
>$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i \longrightarrow \text{Hablar de las ventajas que sea recursiva}$$

# SOLUCIÓN ESTUDIANTE 1

$\frac{p_{i+1}}{p_i} = \frac{n! p^{i+1}(1-p)^{n-i-1}}{(i+1)! (n-i-1)! }\frac{i!(n-i)!}{n! p^{i} (1-p)^{n-i}}$ 

$n!$ se simplifica con $n!$

$\frac{i!}{i+1!}$ se simplifica y queda $\frac{1}{i+1}$ 

$\frac{p^{i+1}}{p^{i}}$ se simplifica y queda $\frac{p}{1}$ 

$\frac{(1-p)^{n-i-1}}{(1-p)^{n-1}}$ se simplifica y queda $(1-p)^{-1}$, es decir, $\frac{1}{(1-p)}$

$\frac{n-i!}{n-i-1!}$ se simplifica y queda $\frac{n-i}{1}$ 

Después de hacer la simplificación nos queda esto

$\frac{p_{i+1}}{p_i} = \frac{(n-i)p}{ (i+1)(1-p)}$ 
$$p_{i+1}=\frac{n-i}{i+1}\frac{p}{1-p} p_i$$

Computacionalmente la carga es mucho menor si lo hacemos de manera recursiva, esto facilita todos los procesos.

# SOLUCIÓN ESTUDIANTE 2

Las ventajas que la función sea recursinva es que no es necesario definir la secuencia de pasos exactos para resolver el problema. La ecuación seria de menor tamaño. De esta manera se conoce la solución explicita, el cual equivaldria a resultado de la ecuación anterior.

# SOLUCIÓN ESTUDIANTE 1

$\frac{p_{k+1}}{p_k} = \frac{e^{-\lambda}(\lambda)^{k+1}}{k+1!} \frac{k!}{e^{-\lambda}(\lambda)^{k}}$ 

$e^{-\lambda}$ se simplifica con $e^{-\lambda}$

$\frac{k!}{k+1!}$ se simplifica y queda $\frac{1}{k+1}$  

$\frac{(\lambda)^{k+1}}{(\lambda)^{k}}$ se simplifica y queda $\frac{\lambda}{1}$ 

Después de hacer la simplificación nos queda esto

$\frac{p_{k+1}}{p_k}=\frac{\lambda}{k+1}$

$$P(k+1)=\frac{\lambda}{k+1}P(k)$$

```python
# Función que calcula la probabilidad acumulada optimizada
def Poisson (n,l):
    Pr = np.zeros(n)
    Pr[0] = np.exp(-l)
    def pr(k):
        nonlocal Pr
        Pr[k+1]=(l/(k+1))*Pr[k]
        
    # Lleno el vector Pr usando compresión de listas
    [pr(k) for k in range(n-1)]
    return np.cumsum(Pr)
```

usando el método de aceptación y rechazo

# SOLUCIÓN ESTUDIANTE 1

```python
# Función de aceptación y rechazo usando compresión de listas
def Acep_rechazo3(N: 'No. variables generadas',
                 f:'función objetivo a generar',
                 t:'función que mayora a f'):
    values=np.zeros(N) 
    def a(i):
        nonlocal values
        R1=np.random.rand()
        R2=np.random.rand()
        f_x = f(R1)
        t_x = t 
        condition = R2*t_x <=f_x 
        while condition!=True: #Si no se cumple la condición repetimos
            R1=np.random.rand()
            R2=np.random.rand()
            f_x = f(R1)
            t_x = t
            condition = R2*t_x <=f_x 
            
        values[i] = R1 
    [a(i) for i in range(N)]
    return values
```

```python
N=10000
x = np.arange(0,1,0.01)
# Función objetivo 
f = lambda x:20*x*(1-x)**3
# Máximo de la función f
t = f(optimize.fmin(lambda x:-f(x),0,disp=False))

X=Acep_rechazo3(N,f,t)
plt.hist(X, density=True, bins=30, label="X generada ")
plt.plot(x, f(x), label="$𝑓(𝑥)=20𝑥(1−𝑥)^3$")
plt.legend()
plt.show()
```
